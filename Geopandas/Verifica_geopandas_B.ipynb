```python
Autostrade = geopandas.read_file("/content/drive/MyDrive/Autostrade_10000_CT10_line.zip")
Autostrade
```

```python
Autostrade['Lunghezza'] = Autostrade['geometry'].length
ax = Autostrade.to_crs(epsg=3857).plot(column = "Lunghezza", cmap='Reds')
ctx.add_basemap(ax)
```

```python
valore = input('inserire valore')
if valore <= 0:
  print('valore negativo')
else:
  max_auto =Autostrade[Autostrade['Lunghezza']>valore]
  ax = valore.to_crs(epsg=3857).plot()
  max_auto.to_crs(epsg=3857).plot(ax=ax,edgecolor='grey')
  ctx.add_basemap(ax)
```

```python
comuni = geopandas.read_file('/content/drive/MyDrive/Limiti01012021_g/Com01012021_g')
comuni
```

```python
comune_input= input()
comu_inserito = comuni[comuni.COMUNE == comune_input]
comuni_limitrofi = comuni[comuni.touchese(comune_input.geometry.squeeze())]
ax= comu_inserito.to_crs(epsg=3857).plot(facecolor='b')
comuni_limitrofi.to_crs(epsg=3857).plot(ax=ax, facecolor='yellow')
ctx.add_basemap(ax)
```
