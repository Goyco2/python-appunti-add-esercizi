```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify    #se non si tropva bisogna scaricarlo 
```

1. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.


```python
d = {'last_name': ['Boyer', 'Copeland', 'Daniel','Serrano', 'Terrell', 'Vargas'],
      'order_numbers' : [185, 50, 196, 58, 24, 28]
     }
df = pd.DataFrame(d)
df
```

2. visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:


```python
data = list(df['order_numbers'])
names = list(df['last_name'])
fig = plt.figure(figsize=(10,10))    #figsize serve  a inserire la misuara del grafico
ax = plt.axes()
ax.bar(names, data)
```

3. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli assi.


```python
data = list(df['order_numbers'])
names = list(df['last_name'])
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
ax.bar(names, data, color ='r')
ax.set_xlabel('last_name')        #set_xlabel serve a inserire il nome della linea
ax.set_ylabel('Order numbers')
fig.suptitle('ISTOGRAMMA')
plt.show()
```

4. Visualizzare lo stesso grafico ma con le barre in ordine decrescente


```python
fig = plt.figure(figsize = (8,7))
ax = plt.axes()
ax.set_xlabel('last_name', fontsize = 20)
ax.set_ylabel('order_numbers', fontsize = 20)
fig.suptitle('ISTOGRAMMA', fontsize = 30)

ax.bar(df.sort_values("order_numbers", ascending = False)["last_name"], df["order_numbers"].sort_values(ascending = False), color = "c")
plt.show()
```

5. visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:


```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
ax.barh(names, data, color ='r')       #barh diversamente da bar lo mette capovolto
ax.set_xlabel('last_name')
ax.set_ylabel('Order numbers')
fig.suptitle('ISTOGRAMMA')
plt.show()
```

6. crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
d = {'last_name': ['Copeland', 'Serrano', 'Boyer','Daniel', 'Vergas', 'Terrell'],
      'order_perc' : [9.242144, 10.720887, 34.195933, 36.229205, 5.175601, 4.436229]
     }
df = pd.DataFrame(d)
df
```

7. visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:


```python
names = list(df['last_name'])
values = list(df['order_perc'])
fig = plt.figure(figsize=[10,10])

ax = plt.axes()
ax.pie(values, labels=names, autopct='%1.2f%%')       #2f%% serve a vedere 2 numeri dopo il punto
plt.show
```

8. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico


```python
names = list(df['last_name'])
values = list(df['order_perc'])
col = ['c','m','r','g','y','b']
fig = plt.figure(figsize=[7,7])

fig.suptitle('grafico a torta', color = 'r')    #titolo
ax = plt.axes()
ax.pie(values, labels=names,colors = col, autopct='%1.2f%%')
plt.show
```

9. rappresentare le stesse informazioni utilizzando come tipo di grafico il TreeMap (google it)


```python
taglia=list(df['order_perc'])
label1 = list(df['last_name'])
squarify.plot(sizes=taglia, label=label1, alpha=0.6 )     #grafico treemap
plt.axis('off')
plt.show()
```
